```python
#import libraries 
import numpy as np
import matplotlib.pyplot as plt 
import mglearn
#import the KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Here we are using **Forge Dataset** .
**A Synthetic two class classification** dataset,which has 2 features.

```python
#generate dataset
X , y = mglearn.datasets.make_forge()
print X.shape
print y.shape
```

```python
#print X[:,0] - get first feature of all row 
#print X[:,1] - get second feature of all row 
#print X - print dataset inputs 
#print y - corresponding label for each inputs in X
```

```python
#We can plot our dataset 
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.legend(["Class 0 ","Class 1"],loc =4)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
plt.show()
```

In the above plot 3 stars are the new data point . 
First we consider only one neighbour and classified to class 0 and 1 . 
Now we will consider 3 neighbors . 

```python
#Considering 3 neighbours
mglearn.plots.plot_knn_classification(n_neighbors=3)
plt.show()
```

Now the prediction of top left point (star) is changed to **Class 1** When we consider only one neighbour it was **Class 0** .

**Another example**

```python
from sklearn.datasets import load_breast_cancer
data_cancer = load_breast_cancer()
```

```python
print data_cancer.DESCR
```

```python
print data_cancer.feature_names
print data_cancer.target_names
print data_cancer.feature_names.shape
print data_cancer.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(data_cancer.data , data_cancer.target,random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#train the model 
clf.fit(X_train,y_train)
```

```python
#Test with test dataset
print clf.predict(X_test)
```

```python
print y_test
```

```python
print "Score : ",clf.score(X_test,y_test)
```

```python
#Test accuracy  
test_accuracy_list = [] 
#training accuracy 
training_accuracy_list = [] 
#No of neighbours 
neighbours = range(1,11)
for n in neighbours:
    #Here n is the number of neighbors
    clssifier = KNeighborsClassifier(n_neighbors=n) 
    #train the model 
    clssifier.fit(X_train,y_train)
    #Append the score of accuracy for training set into the list 
    training_accuracy_list.append(clssifier.score(X_train,y_train))
    #Append the score of accuracy for test set into the list 
    test_accuracy_list.append(clssifier.score(X_test,y_test))

print "Training Accuracy score list : ", training_accuracy_list
print "\nTest Accuracy score list :",test_accuracy_list
```

```python
#Now we will plot this data against number of neighbours 
plt.xlabel("No of Neighbours")
plt.ylabel("Accuracy")
plt.plot(neighbours,training_accuracy_list,label = "training_accuracy")
plt.plot(neighbours,test_accuracy_list,label="test_accuracy")
plt.legend()
plt.show()
```
